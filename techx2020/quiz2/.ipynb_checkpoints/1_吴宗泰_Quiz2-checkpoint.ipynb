## 1. Array Basics
a. 创建一个shape为(10,10), 值为0~99的2D array，命名为x (应该和下图一样)

(Score: 2/2)

```python
x=np.arange(0,100).reshape(10,10)
```

b. 用array slicing找出x中所有5的倍数，并输出为一维array

(Score: 2/2)

```python
a=x[:,:1]
b=x[:,5:6]
c=np.hstack([a,b]).reshape(1,20).flatten()
c
```

c. 找出array x 右上角 shape = (4, 4) 的方块和左下角shape = (2,8)的长方形，把 vertically 他们拼接成一个 shape = (4, 8)的矩阵。把这个矩阵赋值为 y。

(Score: 2/2)

```python
a=x[0:4,-4:].reshape(4,-1)
b=x[-2:,0:8].reshape(4,-1)
y=np.vstack([a,b])
y
```

d. 给y的每个值都apply这个function: $$f(x) = 3 \cdot x + 7$$

```python
3*y+7 
```

e. 创建一个和array x的shape相同的boolean mask, 每个位置的值为array x对应位置的值是否即使3的倍数又是5的倍数。把这个array称作z. (use one of / % //)

(Score: 2/2)

```python
z=(x[:,:]%3==0)&(x[:,:]%5==0)
```

```python
z.shape
```

f. 通过z来选中x中符合这些条件的数。

(Score: 2/2)

```python
print(x[z])
```

f. 使用groupby等方法，计算2008年以及以后每个州18岁以下人口的平均值，输出前五个人口最多的（不包括USA，可以使用 df.sort_values()等函数 ）。

(Score: 1/2)

```python
# "每个州18岁以下人口"

mask=df['year']>=2008
df[mask].groupby('state/region').aggregate(np.mean).sort_values(by=['population'],ascending=False).iloc[1:6]
```

g. 找到每年总人口的maximum，minimum，以及median (用aggreagate)

(Score: 1/2)

```python
# ^

df.groupby('year').aggregate([min,np.median,max])
```
